## Examine Summary Statistics

#### Create a Class with Methods to Examine Summary Statistics

```python
class patient_info:
    # assign columns to class attributes
    def __init__(self, input_df):
        self.age = input_df['age']
        self.sex = input_df['sex']
        self.bmi = input_df['bmi']
        self.children = input_df['children']
        self.smoker = input_df['smoker']
        self.region = input_df['region']
        self.charges = input_df['charges']
    
    # method to analyze summary statistics for age
    def analyze_age(self):
        average = self.age.mean()
        median = self.age.median()
        min_val = self.age.min()
        max_val = self.age.max()
        std_dev = np.std(self.age)
        iqr_val = iqr(self.age)
        print(f"Average age: {average}")
        print(f"Median age: {median}")
        print(f"Age range: {min_val} - {max_val}")
        print(f"Standard deviation: {std_dev}")
        print(f"IQR: {iqr_val}")
        print("")
        
    # method to analyze summary statistics for sex
    def analyze_sex(self):
        male = 0
        female = 0
        total_patients = len(self.sex)
        for patient in self.sex:
            if patient == 'male':
                male += 1
            else:
                female += 1
        male_prop = male / total_patients
        female_prop = female / total_patients
        print(f"Total males: {male}")
        print(f"Total females: {female}")
        print(f"Proportion of males: {male_prop}")
        print(f"Proportion of females: {female_prop}")
        print("")

    # method to analyze summary statistics for bmi
    def analyze_bmi(self):
        average = self.bmi.mean()
        median = self.bmi.median()
        min_val = self.bmi.min()
        max_val = self.bmi.max()
        std_dev = np.std(self.bmi)
        iqr_val = iqr(self.bmi)
        print(f"Average BMI: {average}")
        print(f"Median BMI: {median}")
        print(f"BMI range: {min_val} - {max_val}")
        print(f"Standard deviation: {std_dev}")
        print(f"IQR: {iqr_val}")
        print("")
        
    # method to analyze summary statistics for children
    def analyze_children(self):
        average = self.children.mean()
        median = self.children.median()
        min_val = self.children.min()
        max_val = self.children.max()
        std_dev = np.std(self.children)
        iqr_val = iqr(self.children)
        print(f"Average number of children: {average}")
        print(f"Median number of children: {median}")
        print(f"Range of number of children: {min_val} - {max_val}")
        print(f"Standard deviation: {std_dev}")
        print(f"IQR: {iqr_val}")
        print("")
        
    # method to analyze summary statistics for smoking status
    def analyze_smoker(self):
        smoker = 0
        nonsmoker = 0
        total_patients = len(self.smoker)
        for patient in self.smoker:
            if patient == 'yes':
                smoker += 1
            else:
                nonsmoker += 1
        smoker_prop = smoker / total_patients
        nonsmoker_prop = nonsmoker / total_patients
        print(f"Total smokers: {smoker}")
        print(f"Total nonsmokers: {nonsmoker}")
        print(f"Proportion of smokers: {smoker_prop}")
        print(f"Proportion of non-smokers: {nonsmoker_prop}")
        print("")
        
    # method to analyze summary statistics for region
    def analyze_region(self):
        southwest = 0
        southeast = 0
        northwest = 0
        northeast = 0
        total_patients = len(self.region)
        for patient in self.region:
            if patient == 'southwest':
                southwest += 1
            elif patient == 'southeast':
                southeast += 1
            elif patient == 'northwest':
                northwest += 1
            else:
                northeast += 1
        sw_prop = southwest / total_patients
        se_prop = southeast / total_patients
        nw_prop = northwest / total_patients
        ne_prop = northeast / total_patients
        print(f"Total patients in the Southwest: {southwest}")
        print(f"Total patients in the Southeast: {southeast}")
        print(f"Total patients in the Northwest: {northwest}")
        print(f"Total patients in the Northeast: {northeast}")
        print(f"Proportion of patients in the Southwest: {sw_prop}")
        print(f"Proportion of patients in the Southeast: {se_prop}")
        print(f"Proportion of patients in the Northwest: {nw_prop}")
        print(f"Proportion of patients in the Northeast: {ne_prop}")
        print("")
        
    # method to analyze summary statistics for charges
    def analyze_charges(self):
        average = self.charges.mean()
        median = self.charges.median()
        min_val = self.charges.min()
        max_val = self.charges.max()
        std_dev = np.std(self.charges)
        iqr_val = iqr(self.charges)
        print(f"Average charges: {average}")
        print(f"Median charges: {median}")
        print(f"Range of charges: {min_val} - {max_val}")
        print(f"Standard deviation: {std_dev}")
        print(f"IQR: {iqr_val}")
        print("")
```

#### Instantiate class and call methods

```python
patient_stats = patient_info(df)

patient_stats.analyze_age()
patient_stats.analyze_sex()
patient_stats.analyze_bmi()
patient_stats.analyze_children()
patient_stats.analyze_smoker()
patient_stats.analyze_region()
patient_stats.analyze_charges()
```

## Correlation Analysis

```python
# create correlation matrix heatmap of age, bmi, number of children, and charges
corr_matrix = df.corr(numeric_only=True)

sns.heatmap(corr_matrix, annot=True, cmap='Purples', vmin=-1, vmax=1)
plt.title('Correlations Between Age, BMI, Children, and Charges')
plt.show()
```

```python
# create boxplot of charges to sex
plt.boxplot([df[df['sex'] == 'male']['charges'], df[df['sex'] == 'female']['charges']],
            labels=['Male', 'Female'])
plt.title('Boxplots of Charges by Gender')
plt.xlabel('Gender')
plt.ylabel('Charges')
plt.show()
```

```python
# create boxplot of charges to smoking status
plt.boxplot([df[df['smoker'] == 'yes']['charges'], df[df['smoker'] == 'no']['charges']],
            labels=['Smoker', 'Non-Smoker'])
plt.title("Distribution of Charges Across Smoker Status")
plt.xlabel('Smoker Status')
plt.ylabel('Charges')
plt.show()
```

```python
# create boxplot of charges to region
plt.boxplot([df[df['region'] == 'southwest']['charges'],
             df[df['region'] == 'southeast']['charges'],
             df[df['region'] == 'northwest']['charges'],
             df[df['region'] == 'northeast']['charges']],
            labels=['Southwest', 'Southeast', 'Northwest', 'Northeast'])

plt.title("Distribution of Charges Across Regions")
plt.xlabel('Region')
plt.ylabel('Charges')
plt.show()
```
